# Stint clustering (unsupervised)

Cluster race stint profiles into strategy groups, using the tested `ml`
helpers (#53). k is chosen by silhouette score. Requires `pip install -e ".[analytics]"`.

```python
from fastf1_analytics import session_laps
from ml import cluster_stints, plot_clusters, race_context_features

laps = session_laps(2023, "Singapore", "R")
features, _ = race_context_features(laps)
features = features[features["StintLength"] >= 5].reset_index(drop=True)  # drop fluke short stints
features.head()
```

```python
result = cluster_stints(features)  # k chosen by silhouette over k=2..6
print(f"k={result.k}  silhouette={result.silhouette}  scores={result.k_scores}")
result.summary
```

```python
plot_clusters(features, result.labels, x="StartLap", y="StintLength");
```
